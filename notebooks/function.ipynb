```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(100)
```

```python
y = np.log(1 + np.exp(0.5*x))
```

```python
plt.plot(x, y)
plt.show()
```

```python
y1 = np.log(1 + np.exp(x))
```

```python
plt.plot(x, y1)
plt.show()
```

```python
a = 100
ieta = 1
y2 = ieta * np.sqrt(1 - np.power(x, 2)/a**2)
plt.plot(x, y2)
plt.show()
```
